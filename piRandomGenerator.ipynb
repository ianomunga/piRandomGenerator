#A Simple Monte Carlo Simulation for Generating pi at Random
There are certain numerical values that will always keep occuring at a fixed frequency and interval the more you count. The fact that this will always be true means that finding these values is deterministic. pi is one of these values. 

Monte Carlo simulations enable the calculation of numbers like pi which in this case, will keep showing up the more random numbers you sample, by modelling their probability of occurence, given a set of specific constraints. 

The constraints we have in this case are a couple of Mathematical intuitions:

* That the area of a circle, Area = $πr^2$
* Considering two pairs of randomly generated numbers, x and y, that will represent the coordinates on our sampled graph, generated exclusively as floats between 0 (inclusive) and 1 (exclusive): [0, 1);
* For a square drawn on the cartesian plane such that the length of each side of the square is exactly 1 and the bottom left corner of the square lies at the origin; 
* Any point generated as part of the random pairs will lie within the area bound by the square
* And those of them that fall within the quarter circle [that is itself, within the square] must satisfy the constraint, <br> $x^2+y^2≤1$
* The area of the quarter circle is thus given by: <br> $\frac{πr^2}{4}$
* but r = 1
* Therefore, 
<br>$\frac{π}{4}$ = $\frac{Points within the circle}{Totalpoints generated}$
* It follows therefore, that; <br> $π = 4$ x $\frac{Points within the circle}{Totalpoints generated}$


The value of π we compute in this way gets closer to the true value if we increase the number of points under consideration since a larger number of points brings us closer to a truly random situation.



```python
#Load the libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Defining the lines of the square:
horiz = np.array(range(100))/100.0
y_1 = np.ones(100)
plt.plot(horiz , y_1, 'b')
vert = np.array(range(100))/100.0
x_1 = np.ones(100)
plt.plot(x_1 , vert, 'b')


#Plotting the random points:
import random
inside = 0
i=1
n=int(input("Enter the total number of points: "))
while (i<=n):
  x = random.random()
  y = random.random()
  if ((x**2)+(y**2))<=1:
    inside+=1
    plt.plot(x , y , 'go')
  else:
    plt.plot(x , y , 'ro')
  i+=1
pi=(4*inside)/n
print ("The value of pi is:")
print(pi)
plt.show()
```

It looks like 1000 points goes over pi by 1, so I'll experiment with a higher number to make it even more random and increase the accuracy even further
